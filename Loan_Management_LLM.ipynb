```python
from openai import OpenAI
client = OpenAI(api_key = 'YOUR_API_KEY')
```

```python
import pandas as pd
data = pd.read_csv('train_u6lujuX_CVtuZ9i (1).csv')
data.head()
```

```python
def create_summary_data(data):
  summary = "The dataset has {} rows and {} columns.".format(data.shape[0], data.shape[1])
  summary+="Columns:\n"
  for col in data.columns:
        summary += f"- {col} (type: {data[col].dtype})\n"
  return summary
```

```python
def ai_agent(user_query,data):
  data_context = create_summary_data(data)
  prompt = f"""You are an Loan Expert AI agent.
  you have been provided with a dataset summary {data_context}
  Now based on user's Question {user_query}
  Think step-by-step. Assume you can access and analyze the dataset like a Data Scientist would using Pandas.

Give a clear, final answer."""


  response = client.chat.completions.create(
      model="gpt-3.5-turbo",
      messages=[
          {"role": "user", "content": prompt}],
          temperature=0.2,
          max_tokens=500
  )

  answer = response.choices[0].message.content
  return answer
```

```python
print("Welcome to Loan Review AI Agent!")
print("You can ask anything about the loan applicants data.")
print("Type 'exit' to quit.")
while True:
  user_input = input("\nYour question: ")
  if user_input.lower() == "exit":
        break
  response = ai_agent(user_input, data)
  print("\nAI Agent Response:")
  print(response)
```
